```python
# find output/ -type f -name "*.csv" -newermt 2020-09-17 -not -path "*/.virtfs_metadata/*" | xargs wc -l
import os
import os.path
from os import path
import pandas as pd
from stat import S_ISREG, ST_CTIME, ST_MODE
import time
from datetime import datetime
import re

 
dfs = []
rootDir = 'output3'

def contains_csvfile(fileList):
    for file in fileList:
        if ".csv" in file:
            return True
    return False

def get_csvfile(fileList):
    for file in fileList:
        if ".csv" in file:
            return file
    return None

for dirName, subdirList, fileList in os.walk(rootDir, topdown=False):
    if ".virtfs_metadata" in dirName:
        continue
    
    if not contains_csvfile(fileList):
        print("no CSV file found in", dirName)
        continue
        
    else:        
        
        csv_fname = get_csvfile(fileList)
        
        ### deprecated / unused
        # print('%s\t%s' % (dirName, fname))
        stat_file = os.stat(os.path.join(dirName, csv_fname))
        cdate = stat_file[ST_CTIME]
        creationtime_file = time.ctime(os.path.getmtime(os.path.join(dirName, csv_fname)))
        creationdate_file = datetime.fromtimestamp(os.path.getmtime(os.path.join(dirName, csv_fname))).date()
        targeted_date = datetime.strptime("2020-09-16", "%Y-%m-%d")
        # if creationdate_file.day != targeted_date.day:
        #    continue
        
        def is_file_empty(file_path):
                """ Check if file is empty by confirming if its size is 0 bytes"""
                # Check if file exist and it is empty
                return os.path.exists(file_path) and os.stat(file_path).st_size == 0
        if is_file_empty(os.path.join(dirName, csv_fname)):
            # print("empty CSV file")
            continue
        df = pd.read_csv(os.path.join(dirName, csv_fname))
        df['sampler'] = csv_fname
        
        DEFAULT_BARBARIK_SAMPLER = 9 # KUS is the default sampler
        # processing metainformation (README)
        if path.isfile(os.path.join(dirName, 'README.md')):
            with open(os.path.join(dirName, 'README.md'), 'r') as readme:
                content_readme = readme.read()
                # print(content_readme)
                m = re.search('ref_sampler=(\d+)', content_readme)
                if m is not None:
                    r_sampler = int(m.group(1))
                    print("ref sampler 6 found in", dirName)
                else:
                    r_sampler = DEFAULT_BARBARIK_SAMPLER
                    # print("default sampler is used... KUS", default_sampler)
                    # continue

                df['ref_sampler'] = r_sampler
    
        dfs.append(df)
        
        
        
        
        
  
            
        
            

# for df in dfs:
    # df['sampler'] = solver
#    if len(df['file']) == 1 and "FM-3" in df['file'][0]:
#        print(solver)
#        print(df['file'])
#        print(df['Uniform'].value_counts(dropna=False))
#        print()


all_results = pd.concat(dfs)

all_results.to_csv('all-results-sampler.csv')
all_results[['file', 'time', 'Uniform', 'Timeout', 'sampler']].to_csv('all-results-sampler-digest.csv')
```

```python
def gen_starting_script():
    return '#!/bin/bash\ntimeout=5600\n\n'


SAMPLER_UNIGEN = 1
SAMPLER_QUICKSAMPLER = 2
SAMPLER_STS = 3
SAMPLER_CMS = 4
SAMPLER_UNIGEN3 = 5
SAMPLER_SPUR = 6
SAMPLER_SMARCH = 7
SAMPLER_UNIGEN2 = 8
SAMPLER_KUS = 9
SAMPLER_DISTAWARE = 10

def get_samplerid(sampler):
    if "DistanceBasedSampling" in sampler:
        return SAMPLER_DISTAWARE
    elif "QuickSampler" in sampler:
        return SAMPLER_QUICKSAMPLER
    elif "KUS" in sampler:
        return SAMPLER_KUS
    elif "SPUR" in sampler:
        return SAMPLER_SPUR 
    elif "SMARCH" in sampler:
        return SAMPLER_SMARCH 
    elif "UNIGEN2" in sampler:
        return SAMPLER_UNIGEN2
    elif "UniGen3" in sampler:
        return SAMPLER_UNIGEN3
    elif "UniGen" in sampler:
        return SAMPLER_UNIGEN
    elif "CustomSampler" in sampler:
        return SAMPLER_CMS # CMS?
    elif "STS" in sampler:
        return SAMPLER_STS
    else:
        return -1
    
    
def gen_sampling_script(sampler, bench):
    sampler_id = get_samplerid(sampler)
    if sampler_id == -1:
        print("Unrecognized sampler...", sampler)
        return
    return '''
bench="%s"
sampler=%s
echo "submit with this sampler:"
echo "$sampler"
sed -e "s;SAMPLER_NUMBER;$sampler;g" -e "s;BENCH;$bench;g" -e "s;TIMEOUT;$timeout;g" /udd/macher/usampling/usampling-uniformity-template.sh > /udd/macher/usampling/usampling-uniformity-oarsub$sampler.sh
chmod +x /udd/macher/usampling/usampling-uniformity-oarsub$sampler.sh
oarsub -S /udd/macher/usampling/usampling-uniformity-oarsub$sampler.sh
    ''' % (bench, sampler_id)

first_half = False # first/second half of the bench or all if None
script = gen_starting_script()
for s in all_results.query('ref_sampler == 6')["sampler"].unique():
    print("for sampler=", s, get_samplerid(s))
    if get_samplerid(s) == 6: # ref sampler with sampler has no sense for SPUR
        continue
    bench_proc = all_results.query("ref_sampler == 6 and sampler == '" + s + "'")['file'].unique()
    print(len(bench_proc))    
    # print('bench="' + " ".join(list(set(bench.split(" ")) - set(bench_proc))) + '"')
    diff_bench = list(set(bench.split(" ")) - set(bench_proc))    
    # first_half or second half or both
    if first_half is not None:
        length_diff = len(diff_bench)
        middle_index = length_diff//2
        if first_half:
            diff_bench = diff_bench[:middle_index]
        else: # second half
            diff_bench = diff_bench[middle_index:]
    print("missing", len(diff_bench))
    if len(diff_bench) > 0:
        script = script + gen_sampling_script(s, " ".join(diff_bench)).strip()
    else:
        print("nothing to do for this sampler")
    script = script + '\n\n'
print(script)

with open("usampling-uniformity-genresume.sh", "w") as bash_script:
    bash_script.write(script)
```
